```python
# get dataset from URL API Ayuntamiento Madrid
def get_dataset(url):
    # get response of url and convert to json
    response = requests.get(url)
    dataset_json = response.json()
    return dataset_json  
```

```python
# join two datasets (both are list of json)
def join_datasets(dt1, dt2):
    # join datasets (list of dict)
    dt_join = dt1 + dt2
    return dt_join
```

```python
# get location in datset for a name of place. Retunr a list if there is some place with same name
def get_place(name, data):
    # get first place found 
    place_list = [d for d in data if d["title"] == name]
    if len(place_list) == 0:
        return None
    else:
        return place_list[0]
```

```python
# get name for a place
def get_name(place):
    # get name
    name_place = place["title"]
    return name_place
```

```python
# get location for a place
def get_location(place):
    # get location
    loc_place = place["location"]
    return loc_place

# get latitude for a location
def get_latitude(location):
    # get latitude
    lat = location["latitude"]
    return lat

# get longitude for a location
def get_longitude(location):
    # get longitude
    lon = location["longitude"]
    return lon
```

```python
DATASET_CULTURAL = "https://datos.madrid.es/egob/catalogo/200304-0-centros-culturales.json"
DATASET_MUSEUM = "https://datos.madrid.es/egob/catalogo/201132-0-museos.json"

my_dataset = join_datasets(get_dataset(DATASET_CULTURAL)["@graph"], get_dataset(DATASET_MUSEUM)["@graph"])

print(len(my_dataset))

my_df = pd.DataFrame(my_dataset)
print(my_df.count())
my_df.head(5)
```

```python
my_interest_place = "Casita - Museo del Ratón Pérez"

# get place
my_place = get_place(my_interest_place, my_dataset)
if my_place == None:
    print("placen not found")
else:
    # get name place 
    my_name_place = get_name(my_place)
    # get location place 
    my_location = get_location(my_place)
    # get latitude place 
    my_lat = get_latitude(my_location)
    # get longitude place 
    my_lon = get_longitude(my_location)

    print(f"{my_name_place} is in latitude {my_lat} and longitude {my_lon}")
```
