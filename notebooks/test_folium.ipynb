```python
import folium
```

```python
m = folium.Map(location=[40.965, -5.664])
m
```

```python
loc_1 = [40.44342, -3.72693]
loc_2 = [40.417048, -3.705529]

my_map = folium.Map(location=loc_1,
                    zoom_start=13)

folium.Marker(loc_1, tooltip="title loc 1", icon=folium.Icon(color="red", prefix="fa", icon="bicycle")).add_to(my_map)
folium.Marker(loc_2, tooltip="title loc 2", icon=folium.Icon(color="green")).add_to(my_map)

my_map
```

```python
my_map.save("../results/map.html")
```
